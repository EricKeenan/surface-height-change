# This notebook calculates and saves daily sums from MERRA-2 NetCDF files
# Note that the EVAP record is broken. So SMB only includes PRECIP

```python
import xarray as xr
import os
```

```python
def calc_daily_sum(src_path, tgt_path):
    
    ds = xr.open_dataset(src_path)
    ds = ds.sortby('time', ascending=True)
    daily_sum = ds.resample(time='D').mean(dim='time')
    daily_sum.to_netcdf(path=tgt_path)    
```

```python
src_dir = "/scratch/summit/erke2265/MERRA2/"
```

# PRECSN

```python
# Settings
tgt_var = "PRECSN"

for filename in os.listdir(src_dir):
    if tgt_var in filename:
        src_path = src_dir + filename
        tgt_path = "/scratch/summit/erke2265/surface-height-change/data/M2_" + filename
        tgt_path = tgt_path.replace("hourly", "daily")  
        print(tgt_path)
        calc_daily_sum(src_path, tgt_path)
        
ds = xr.open_mfdataset("/scratch/summit/erke2265/surface-height-change/data/M2_" + tgt_var + "_daily_*.nc", combine='nested', concat_dim='time')
ds = ds.sortby('time', ascending=True)
ds.to_netcdf(path="/scratch/summit/erke2265/surface-height-change/data/M2_" + tgt_var + "_daily_1980-2020.nc")
```

# SMB

```python
sn = xr.open_dataset("/scratch/summit/erke2265/surface-height-change/data/M2_PRECSN_daily_1980-2020.nc")['PRECSN']
cu = xr.open_dataset("/scratch/summit/erke2265/surface-height-change/data/M2_PRECCU_daily_1980-2020.nc")['PRECCU']
ls = xr.open_dataset("/scratch/summit/erke2265/surface-height-change/data/M2_PRECLS_daily_1980-2020.nc")['PRECLS']
evap = xr.open_dataset("/scratch/summit/erke2265/surface-height-change/data/M2_EVAP_daily_1980-2020.nc")['EVAP']
```

```python
smb = sn + ls + cu - evap
smb = smb * 60 * 60 * 24
smb
```

```python
smb.to_netcdf(path="/scratch/summit/erke2265/surface-height-change/data/M2_1980-2020_daily_SMB.nc")
```
